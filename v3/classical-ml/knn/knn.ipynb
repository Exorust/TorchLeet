### Problem Statement

K-Nearest Neighbors (KNN) is a simple yet powerful classification algorithm. Given a query point, it finds the K closest points in the training set and predicts the most common label among them (majority vote).

Your task is to implement a KNN classifier **from scratch** using PyTorch tensor operations.

---

### Requirements

1. **Distance Computation** — Compute Euclidean distances between query points and all training points.
2. **Neighbor Selection** — Find the K nearest neighbors using `torch.topk`.
3. **Majority Voting** — Predict the class label by majority vote among the K neighbors using `torch.mode`.
4. **Batch Support** — The function should classify multiple query points at once.

---

### Constraints

- ✅ Use only PyTorch tensor operations.
- ✅ Use `torch.cdist` for efficient distance computation.
- ✅ Use `torch.topk` for finding nearest neighbors (not sorting the entire distance matrix).
- ❌ Do **not** use sklearn or any external ML libraries.

---

<details>
  <summary>💡 Hint</summary>

  - `torch.cdist(X_test, X_train)` gives you a `(num_test, num_train)` distance matrix.
  - `torch.topk(distances, k, largest=False)` returns the K smallest distances and their indices.
  - Use the indices to gather the labels: `train_labels[indices]` gives you a `(num_test, K)` tensor of neighbor labels.
  - `torch.mode(neighbor_labels, dim=1)` returns the most frequent label per row.

</details>
